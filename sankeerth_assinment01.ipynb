```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df=pd.read_csv('sales_data_with_discounts.csv')
```

```python
numerical_columns=df.select_dtypes(include=[np.number])
```

```python
print(numerical_columns)
```

```python
mean_values = numerical_columns.mean()
```

```python
print(mean_values)
```

```python
median_values = numerical_columns.median()
```

```python
mode_values = numerical_columns.mode().iloc[0]
```

```python
print(mode_values)
```

```python
std_dev_values = numerical_columns.std()
```

```python
print(std_dev_values)
```

```python
# plat historams for numerical columns
for column in numerical_columns.columns:
  plt.figure(figsize=(10,8))
  sn.histplot(data=df,x=column,kde=True)
  plt.title('Historam of numerical column')
  plt.show()
```

```python
skewness=numerical_columns[column].skew()
```

```python
q1=numerical_columns[column].quantile(0.25)
q3=numerical_columns[column].quantile(0.75)
iqr=q3-q1
lower_bound=q1-1.5*iqr
upper_bound=q3+1.5*iqr
```

```python
print(skewness)
```

```python
print(f"potential outliers: values below{lower_bound} or above {upper_bound}\n")
```

```python
# Create boxplots for numerical variables
plt.figure(figsize=(10, 8))
df.boxplot()
plt.title('Boxplot of Numerical Variables')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
df.describe()
```

```python
 outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)][column]
```

```python
 print(f"Potential outliers: {outliers.tolist()}")
```

```python
categorical_columns = df.select_dtypes(include=['object']).columns.tolist()
```

```python
categorical_columns
```

```python
# barchart for categorical columns
for col in categorical_columns:
    plt.figure(figsize=(8, 5))
    df[col].value_counts().sort_index().plot(kind='bar', color='skyblue')
    plt.title(f'Count of Categories in {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.grid(axis='y')
    plt.show()
```

```python
# Analyzing the distribution of categories
for col in categorical_columns:
    print(f"Column '{col}':")
    print(df[col].value_counts())
    print()
```

```python
# Function to standardize numerical columns
def standardize_column(column):
    mu = np.mean(column)
    sigma = np.std(column)
    return (column - mu) / sigma
```

```python
for col in df.columns:
    if np.issubdtype(df[col].dtype, np.number):
        df[col + '_standardized'] = standardize_column(df[col])
        plt.figure(figsize=(14, 6))
```

```python
for i, col in enumerate(df.columns):
    if col.startswith('Variable'):
        plt.subplot(1, 3, i + 1)
        plt.hist(df[col], bins=8, color='lightblue', edgecolor='black')
        plt.title(f'Original Distribution of {col}')
        plt.xlabel('Value')
        plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 6))
```

```python
# Plot standardized distributions
for i, col in enumerate(df.columns):
    if col.endswith('_standardized'):
        plt.hist(df[col], bins=8, color='lightgreen', edgecolor='black')
        plt.title(f'Standardized Distribution of {col[:-13]}')
        plt.xlabel('Standardized Value')
        plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Apply one-hot encoding to categorical columns
df_encoded = pd.get_dummies(df, columns=['Date', 'Day', 'SKU', 'City', 'BU', 'Brand', 'Model'])
```
